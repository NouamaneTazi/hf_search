```python
import os
import pandas as pd
import json
from tqdm.autonotebook import tqdm
df = []
for root, dirs, files in tqdm(os.walk('models')):
    if len(files) > 0:
        infos = json.load(open(os.path.join(root, "infos.json"), 'r'))
        if "README.md" in files:
            try:
                infos["readme"] = open(os.path.join(root, "README.md"), 'r').read()
            except:
                print("Could not read README.md from file " + os.path.join(root, "README.md"))
        df.append(infos)
df = pd.DataFrame(df)
df.to_json('models.jsonl', orient='records', lines=True)
df
```
